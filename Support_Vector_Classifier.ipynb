```python
# Importing necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Load the dataset
file_path = '/content/Auto.csv'
auto_df = pd.read_csv(file_path)

# Identify features for scaling (excluding non-numeric and target variables)
features_to_scale = ['cylinders', 'displacement', 'weight', 'acceleration', 'year']

# Scale features to mean 0 and standard deviation 1
scaler = StandardScaler()
auto_df[features_to_scale] = scaler.fit_transform(auto_df[features_to_scale])

# Create binary variable for 'mpg' above/below median
mpg_median = auto_df['mpg'].median()
auto_df['high_mpg'] = (auto_df['mpg'] > mpg_median).astype(int)

# Define feature set (excluding 'mpg' and non-numeric columns) and target variable
X = auto_df.drop(columns=['mpg', 'name', 'origin', 'high_mpg', 'horsepower'])  # Exclude mpg and non-relevant columns
y = auto_df['high_mpg']

# List of different values for the regularization parameter C
C_values = [0.01, 0.1, 1, 10, 100]
cross_val_errors_linear = {}

# Perform cross-validation for each C value with a linear kernel
for C in C_values:
    svc_linear = SVC(C=C, kernel='linear', random_state=0)
    scores_linear = cross_val_score(svc_linear, X, y, cv=5, scoring='accuracy')
    error_linear = 1 - scores_linear.mean()  # Calculate cross-validation error
    cross_val_errors_linear[C] = error_linear

# Display the cross-validation errors for each value of C using a linear kernel
cross_val_errors_linear
```

The cross-validation errors for the support vector classifier with a linear kernel and different C values are:-

*   C = 0.01: 0.1058
*   C = 0.1: 0.153

*   C = 1: 0.138
*   C = 10: 0.138

*   C = 100: 0.1307

Analysis:-

*   Lower values of C tend to have higher cross-validation errors, indicating underfitting.
*   Higher values of C slightly reduce error, but there is minimal performance gain beyond a certain point.

*   The classifier's performance does not improve significantly with very large C values, suggesting that further increasing C may lead to diminishing accuracy.





```python
from sklearn.metrics import make_scorer, f1_score

# Define custom scorer using F1 score
f1_scorer = make_scorer(f1_score)

# Dictionary to store F1 score cross-validation errors for each value of C
cross_val_f1_errors = {}

# Perform cross-validation for each C value using the F1 score
for C in C_values:
    svc_linear = SVC(C=C, kernel='linear', random_state=0)
    scores_f1 = cross_val_score(svc_linear, X, y, cv=5, scoring=f1_scorer)
    error_f1 = 1 - scores_f1.mean()  # Calculate F1-based cross-validation error
    cross_val_f1_errors[C] = error_f1

# Display the F1 score cross-validation errors for each value of C
cross_val_f1_errors
```

The F1-based cross validation errors for the support vector classifier with a linear kernel and various values of C are:-

*   C = 0.01: 0.1035
*   C = 0.1: 0.189

*   C = 1: 0.179
*   C = 10: 0.175

*   C = 100: 0.157

Analysis:-

*   With the F1 score as the evaluation metric, we observe a similar trend to the accuracy-based evaluation:
       

*   Lower values of C yield the lowest error, indicating relatively good performance on balanced precision and recall.
*   Higher values continue to show slight improvements, although performance gains diminish as C increases beyond a certain point.

These results suggest that C = 0.01 might still be a balanced choice, especially if maximizing the balance between false positives and false negatives is critical.

```python
# Define new parameter values for gamma in radial basis function (RBF) kernel and degree for polynomial kernel
gamma_values = [0.01, 0.1, 1]
degree_values = [2, 3, 4]

# Dictionaries to store F1 score cross-validation errors for each kernel and parameter configuration
cross_val_f1_errors_rbf = {}
cross_val_f1_errors_poly = {}

# Evaluate SVM with radial basis function (RBF) kernel using different values of C and gamma
for C in C_values:
    for gamma in gamma_values:
        svc_rbf = SVC(C=C, kernel='rbf', gamma=gamma, random_state=0)
        scores_f1_rbf = cross_val_score(svc_rbf, X, y, cv=5, scoring=f1_scorer)
        error_f1_rbf = 1 - scores_f1_rbf.mean()  # Calculate F1-based cross-validation error
        cross_val_f1_errors_rbf[(C, gamma)] = error_f1_rbf

# Evaluate SVM with polynomial kernel using different values of C and degree
for C in C_values:
    for degree in degree_values:
        svc_poly = SVC(C=C, kernel='poly', degree=degree, random_state=0)
        scores_f1_poly = cross_val_score(svc_poly, X, y, cv=5, scoring=f1_scorer)
        error_f1_poly = 1 - scores_f1_poly.mean()  # Calculate F1-based cross-validation error
        cross_val_f1_errors_poly[(C, degree)] = error_f1_poly

# Display the F1 score cross-validation errors for each configuration of RBF and polynomial kernels
cross_val_f1_errors_rbf, cross_val_f1_errors_poly
```

The F1-based cross-validation errors for the SVM models with radial basis finction (RBF) and polynomial kernels across various values of C, gamma, and degree are as follows:-

Radial Basis Function (RBF) Kernel:-

*   Optimal Errors:- Lower F1-score errors are achieved with moderate C values and gamma values around 0.1.
*   High Error at Low Gamma:- Small gamma values with low C result in high errors suggesting underfitting.

*   Overfitting:- Higher gamma values, combined with large C values, tend to increase the error slightly, suggesting that an excessively complex boundary leads to overfitting.

Polynomial Kernel:-

*   Degree Effects:- The polynomial kernel shows variability in performance based on the degree. Lower degrees(like 2) generally achievelower F1 errors compared to higher degrees.
*   Optimal Parameters:- Moderate C values combined with degree 2 or 3 offer reasonable performance, with C = 1 and degree 2 achieving an F1-error around 0.55.

*   Complex Boundaries:- Higher degrees or larger C values result in increased error, indicating potential overfitting


Conclusion:-

The RBF kernel generally performs better with moderate C and gamma values. The polynomial kernel can also be effective, but higher degrees do not improve F1-score and may lead to overfitting.
